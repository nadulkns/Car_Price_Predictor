```python
import pandas as pd
```

```python
df=pd.read_csv('carprice.csv')
```

```python
df.head()
```

```python
df1=df.dropna()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df1[~df1['price'].apply(is_float)]
```

```python
df2=df1.drop(df1[~df1['price'].apply(is_float)].index.values,axis='index')
```

```python
df2.head()
```

```python
df2.make.unique()
```

```python
make_stat=df2.groupby('make')['make'].agg('count').sort_values(ascending=False)
```

```python
make_less_5=make_stat[make_stat<=5]
```

```python
make_less_5
```

```python
df3=df2.copy()
```

```python
df3.make=df3.make.apply(lambda x:'other' if x in make_less_5 else x)
```

```python
df3.make.unique()
```

```python
dum1=pd.get_dummies(df3.make)
dum1=dum1.astype(int)
```

```python
dum1
```

```python
dum2=pd.get_dummies(df3['fuel-type'])
dum2=dum2.astype(int)
```

```python
dum3=pd.get_dummies(df3['body-style'])
dum3=dum3.astype(int)
```

```python
df3['fuel-type'].unique()
```

```python
df3['body-style'].unique()
```

```python
df4 = pd.concat([df3, dum1,dum2,dum3], axis='columns')
```

```python
df4.head()
```

```python
df5=df4.drop(['make','fuel-type','body-style'],axis='columns')
```

```python
df5.head()
```

```python
X=df5.drop('price',axis='columns')
y=df5.price
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2,shuffle=True)
```

```python
from sklearn.linear_model import LinearRegression

lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train)
score = lr_clf.score(X_test, y_test)
```

```python
score
```

```python
import numpy as np
def predict_price(make, fueltype, bodytype, enginesize,horsepower):
    loc_index_make = np.where(X.columns == make)[0]
    loc_index_fueltype=np.where(X.columns == fueltype)[0]
    loc_index_bodytype=np.where(X.columns == bodytype)[0]

    x = np.zeros(len(X.columns))
    x[0] = enginesize
    x[1] = horsepower
    if loc_index_make >= 0:
        x[loc_index_make[0]] = 1
    if loc_index_fueltype>=0:
        x[loc_index_fueltype[0]] = 1
    if loc_index_bodytype>=0:
        x[loc_index_bodytype[0]] = 1    

    return lr_clf.predict([x])[0]
```

```python
np.round(predict_price('audi','diesel','sedan','180','150'),2)
```

```python
import json
import pickle

with open('car_prices_model.pickle', 'wb') as f:
    pickle.dump(lr_clf, f)

columns = {
    'data_columns': [col.lower() for col in X.columns]
}
with open("columns.json", "w") as f:
    f.write(json.dumps(columns))
```
